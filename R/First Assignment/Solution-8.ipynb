Proof of central limit theorem: 
Sample from any distribution follows normal distribution

```r
population=rchisq(10000,5,0.5)  
hist(population, main="ChiSquare Distribution",breaks=100) 
```

```r
#Sample from a chisquare distribution follows normal distribution
dim(population)=c(10,1000)
sample=colMeans(population) 
hist(sample, main="Normal Distribution",breaks=100)
```
