Write a R function where you can show the proof of Central limit theorem visually. The input should be any PDF and their parameter value. The output should be the graph of sampling distribution. 

```r
#Population follows exponential distribution
population=rexp(10000,5)  
hist(population, main="Exponential Distribution",breaks=80) 
```

```r
#Sample from a exponential distribution follows normal distribution
dim(population)=c(10,1000)
sample=colMeans(population) 
hist(sample, main="Normal Distribution",breaks=80)
```
