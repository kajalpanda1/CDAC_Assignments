Use petrol_consumption dataset. Your task is to predict the gas consumption (in millions of gallons) in 48 of the US states based on petrol tax (in cents), per capita income (dollars), paved highways (in miles) and the proportion of population with the driving license. Build the regression model using KNN Regressor. Analyze the prediction ability of your model.

```
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn import metrics as m
```

```
petro_df=pd.read_csv('https://raw.githubusercontent.com/kajalpanda1/DatasetRepository/master/Petrol_consumption.csv')
```

```
x=petro_df.iloc[:, 0:4].values
y=petro_df.iloc[:, 4].values
```

```
x
```

```
y
```

```
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.15,random_state=10)
```

```
neigh = KNeighborsRegressor(n_neighbors=10)
model=neigh.fit(X_train,y_train)
```

```
#Training Error
y_pred=model.predict(X_train)
MSE=m.mean_squared_error(y_pred,y_train)
print('Mean Squared Error:',MSE)
RMSE=np.sqrt(MSE)
print('Root Mean Squared Error:',RMSE)
```

```
#Testing Error
y_pred1=model.predict(X_test)
MSE=m.mean_squared_error(y_pred1,y_test)
print('Mean Squared Error:',MSE)
RMSE=np.sqrt(MSE)
print('Root Mean Squared Error:',RMSE)
```
