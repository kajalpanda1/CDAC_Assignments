Use Wholesale Customers dataset and perform the following:

a) Implement Pre-processing functions

- Check for NULL values in the dataset. If any NULL value is present then treat it accordingly.

- extract only feature columns from the dataset.

b) Determine the number of clusters using dendogram. Use ward linkage method.

c) Apply agglomerative clustering using ward linkage method.

d) Form the clusters.

e) Visualize the clusters formed by agglomerative clustering.


```
#importing wholesale dataset
import pandas as pd 
import matplotlib.pyplot as plt 
df=pd.read_csv('https://raw.githubusercontent.com/kajalpanda1/DatasetRepository/master/Wholesale_Customers.csv') 
```

```
#checking for NULL values
df.isnull().sum().sort_values(ascending=False) 
```

```
#extracting features from dataset
df_feature=df.iloc[:,2:8].values 
```

```
#standardizing feature values
from sklearn.preprocessing import StandardScaler 
sc=StandardScaler() 
df_scaled=sc.fit_transform(df_feature)
```

```
#plotting dendrogram to determine number of clusters
import scipy.cluster.hierarchy as shc 
plt.figure(figsize=(10,7)) 
plt.title("Wholesale Data Dendrogram") 
dend=shc.dendrogram(shc.linkage(df_scaled,method='ward'))
```

```
#implementing agglomerative clustering
from sklearn.cluster import AgglomerativeClustering 
cluster=AgglomerativeClustering(n_clusters=3,linkage='ward') 
cluster.fit(df_scaled) 
labels=cluster.labels_  
```

```
#visualizing formed clusters
plt.figure(figsize=(10,7)) 
plt.scatter(df_scaled[:,0],df_scaled[:,1],c=cluster.fit_predict(df_scaled),cmap='rainbow') 
```
